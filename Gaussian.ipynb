### Question 1B: Gaussian

```python
import numpy as np
import matplotlib.pyplot as plt

sample = [2,3,4,8,10,11,12]
```

```python
def gauss_func(x,xi,h):
    out1=np.exp(-0.5*((x-xi)**2/h**2))
    #print(out1)
    final1=1/np.sqrt(2*3.14*h)
    #print(final1)
    output1=final1*out1
    #print(output1)
    #output1=int(output1)
    return output1
```

```python
def gaussian(data,h1):
    points1 = np.linspace(min(data)-h1/2,max(data)+h1/2,100)
    final0=[]
    res0=0
    for i in points1:    
        i=round(i,2)
        for itr in data:
            res0=res0+gauss_func(i,itr,h1)
        result1=(1/len(data))*res0 
        res0=0
        final0.append(result1)
    plt.plot(points1,final0)
    plt.show()
    #print(final0)      
    
    
```

```python
gaussian(sample,0.5)
```

```python
gaussian(sample,1)
```

```python
gaussian(sample,1.5)
```
